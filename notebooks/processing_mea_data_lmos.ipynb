```python
import matplotlib.pyplot as plt
def plot_firing_rates(data, bin_size=0.001):
    """
    Plots the firing rates for each channel in a 10x6 grid of subplots.
    
    Parameters:
        data (numpy.ndarray): A 60xN matrix containing the firing rates for 60 channels.
    """

    duration_secs = data.shape[1] / 1000  # Convert number of samples to seconds (assuming 10 kHz sampling rate)
    
    # Time vector in seconds, given 10 kHz sampling rate
    time_vector = np.arange(0, duration_secs, bin_size)
    
    # Create subplots
    fig, axs = plt.subplots(10, 6, figsize=(15, 20))
    
    for i, ax in enumerate(axs.flatten()):
        if i < data.shape[0]:  # Check to avoid IndexError if data has fewer than 60 channels
            ax.plot(time_vector, data[i])
            ax.set_title(f'Channel {i+1}')
            ax.set_xlabel('Time (s)')
            ax.set_ylabel('Firing Rate')
    
    plt.tight_layout()
    plt.show()
    
# run the plot_firing_rates function on the firing rates for the recording name
plot_firing_rates(firing_rates)
```

```python
def get_ordered_event_timestamps_in_ms(file_path):
    """
    Extract and sort event timestamps from an MCS HDF5 file.

    This function reads event data from a Multi Channel Systems (MCS) HDF5 file,
    extracts the timestamps for all recorded events, converts them from
    microseconds to milliseconds, and returns them in chronological order.

    Parameters
    ----------
    file_path : str
        The file path to the MCS HDF5 file containing event data.

    Returns
    -------
    event_timestamps_ms : list of float
        A sorted list of event timestamps in milliseconds. Only positive,
        non-zero timestamps are included, indicating the times at which events
        occurred.

    Notes
    -----
    This function assumes that the HDF5 file follows the structure used by
    MCS data files, with event data located within the first recording and
    first event stream of the file. It also assumes that the timestamps are
    stored as positive integers representing microseconds since the start of
    the recording.

    Examples
    --------
    >>> test_data_folder = '/path/to/your/data/folder'
    >>> file_path = os.path.join(test_data_folder, 'your_data_file.h5')
    >>> timestamps_in_ms = get_ordered_event_timestamps_in_ms(file_path)
    >>> print(timestamps_in_ms)
    [56300.0, 63200.0, 194500.0]
    """
    
    # Load the raw data from the file path
    channel_raw_data = McsPy.McsData.RawData(file_path)
    
    # Get the first event stream for the recording
    event_streams = channel_raw_data.recordings[0].event_streams[0]
    
    # Create the dictionary to store the event streams
    event_entities_dict = event_streams.event_entity
    
    # Create an empty list to store the timestamps for all events in order
    event_timestamps_ms = []

    # Loop over the keys in the event entities dictionary
    for key in event_entities_dict.keys():
        # Get the events for the current key
        events_data = event_entities_dict[key].get_events()
        events_array = events_data[0]  # The first element is the array of timestamps
        
        # Extract the timestamp and convert to milliseconds
        for event in events_array:
            if event[0] > 0:  # Check for a valid timestamp
                timestamp_ms = event[0] / 1000  # Convert microseconds to milliseconds
                event_timestamps_ms.append(timestamp_ms)

    # Sort the timestamps to ensure they are in order
    event_timestamps_ms.sort()

    return event_timestamps_ms


# Use the function like this
test_data_folder = '/home/cresp1el-local/Documents/MATLAB/Data/comparing_lmos_project_folder/SpikeStuff/lmo3_ctz/lmo3_ctz_1'
file_path = os.path.join(test_data_folder, '2022-08-02T15-30-38McsRecording.h5')
timestamps_in_ms = get_ordered_event_timestamps_in_ms(file_path)

# The timestamps_in_ms list now contains the sorted millisecond timestamps
print(timestamps_in_ms)
```

```python
def get_unique_recording_names(df):
    """
    Extracts unique recording names from a DataFrame column.

    Parameters
    ----------
    df : pandas.DataFrame
        The DataFrame that contains the 'recording_name' column.

    Returns
    -------
    unique_names : list of str
        A list of unique recording names as strings.
    
    Examples
    --------
    >>> df = pd.read_csv('path_to_csv_or_excel')
    >>> unique_recording_names = get_unique_recording_names(df)
    >>> print(unique_recording_names)
    """
    # Ensure that the 'recording_name' column exists
    if 'recording_name' not in df.columns:
        raise ValueError("DataFrame does not contain a 'recording_name' column.")
    
    # Get unique names from the 'recording_name' column
    unique_names = df['recording_name'].unique().tolist()
    
    # Convert numpy objects to strings if they are not already
    unique_names = [str(name) for name in unique_names]
    
    return unique_names

# Usage example:
# Assuming you have a DataFrame 'df' loaded with the data
unique_names = get_unique_recording_names(analysis_mea.recording_results_df)
print(unique_names)
```

```python
import os
import glob

def construct_file_paths(df, base_project_path):
    """
    Constructs full file paths for .h5 recording files based on a DataFrame of recording names.

    Parameters
    ----------
    df : pandas.DataFrame
        The DataFrame that contains the 'recording_name' column.
    base_project_path : str
        The base path that is constant for all file paths.

    Returns
    -------
    file_paths : dict
        A dictionary where keys are recording names and values are the full file paths for the .h5 files.
    """
    # Get the unique recording names
    recording_names = get_unique_recording_names(df)

    # Dictionary to store the file paths
    file_paths = {}

    for recording_name in recording_names:
        # Split the recording name on the last underscore to separate the group and the recording index
        parts = recording_name.rsplit('_', 1)
        if len(parts) == 2:
            group_name, _ = parts
            # Construct the folder path for this recording
            folder_path = os.path.join(base_project_path, group_name, recording_name)
            
            # Search for any .h5 file in the folder path
            h5_files = glob.glob(os.path.join(folder_path, '*.h5'))
            
            # Check if there is any .h5 file present
            if h5_files:
                # Assuming we take the first .h5 file if there are multiple
                h5_filename = os.path.basename(h5_files[0])
                print(f"Found .h5 file: {h5_filename}")
                
                # Construct the full path for the .h5 file
                full_path = os.path.join(folder_path, h5_filename)
                file_paths[recording_name] = full_path
            else:
                print(f"No .h5 file found in directory: {folder_path}")
                file_paths[recording_name] = None

    return file_paths

# Usage example:
base_project_path = '/home/cresp1el-local/Documents/MATLAB/Data/comparing_lmos_project_folder/SpikeStuff/'
file_paths = construct_file_paths(analysis_mea.recording_results_df, base_project_path)
print(file_paths)
```

```python
def add_h5_file_paths_to_df(df, file_paths):
    """
    Adds a new column 'h5_file_paths' to the DataFrame with the constructed file paths.

    Parameters
    ----------
    df : pandas.DataFrame
        The DataFrame to which the file paths will be added.
    file_paths : dict
        A dictionary where keys are recording names and values are the full file paths for the .h5 files.

    Returns
    -------
    df : pandas.DataFrame
        The DataFrame with the new 'h5_file_paths' column added.
    """
    # Create a new column with default value as None
    df['h5_file_paths'] = None

    # Iterate over the dictionary and update the DataFrame
    for recording_name, path in file_paths.items():
        df.loc[df['recording_name'] == recording_name, 'h5_file_paths'] = path

    return df


# Now, add the file paths to the DataFrame
analysis_mea.recording_results_df = add_h5_file_paths_to_df(analysis_mea.recording_results_df, file_paths)


analysis_mea.recording_results_df
```

```python
def add_event_timestamps_to_df(df):
    """
    Adds a new column 'events_timestamp_ms' to the DataFrame with the event timestamps.

    Parameters
    ----------
    df : pandas.DataFrame
        The DataFrame that contains the 'h5_file_paths' column.

    Returns
    -------
    df : pandas.DataFrame
        The DataFrame with the new 'events_timestamp_ms' column added.
    """
    # Create a new column for event timestamps
    df['events_timestamp_ms'] = None

    # Iterate over the DataFrame rows
    for index, row in df.iterrows():
        # Check if the h5_file_paths is not None
        if row['h5_file_paths']:
            try:
                # Get the ordered event timestamps in milliseconds for the .h5 file
                event_timestamps = get_ordered_event_timestamps_in_ms(row['h5_file_paths'])
                # Assign the event timestamps to the corresponding cell in the DataFrame
                df.at[index, 'events_timestamp_ms'] = event_timestamps
            except Exception as e:
                # If there's an error, you can decide to print it, pass, or store it in the DataFrame
                print(f"Error processing file {row['h5_file_paths']}: {e}")
                df.at[index, 'events_timestamp_ms'] = None

    return df

# Assuming your DataFrame is named df and it has the 'h5_file_paths' column populated:
analysis_mea.recording_results_df  = add_event_timestamps_to_df(analysis_mea.recording_results_df)

analysis_mea.recording_results_df 
```
